```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
emp = pd.read_csv('emp.csv')
emp
```

```python
dept = pd.read_csv('dept.csv')
dept
```

```python
#exam01) emp.csv를 로드한 후 이름과 봉급을 기점으로 선 그래프를 그린다.
array_ename = np.array(emp['ename'])
array_sal = np.array(emp['sal'])
plt.plot(array_ename, array_sal)
```

```python
#exam02) 직업별로 봉급을 그룹화 한 결과를 원그래프를 그려서 수치를 범례표시한다.
group = emp.groupby('job')['sal'].sum().reset_index()
group
```

```python
g_job = np.array(group['job'])
g_sal = np.array(group['sal'])
plt.pie(g_sal, labels=g_job, autopct='%.1f%%')
```

```python
# exam03) 부서별 결과를 막대그래프로 표시한다. 단, 봉급의 합을 이용한다.
merge = pd.merge(emp, dept, on = 'deptno')
a = merge.groupby('dname')['sal'].sum().reset_index()
a
```

```python
d = np.array(a['dname'])
s = np.array(a['sal'])
plt.bar(d, s)
```

```python
#exam04) 직업별 부서별 두 개의 봉급의 최대값으로 곡선을 그려낸다.
print(s.max())
print(g_sal.max())
```
