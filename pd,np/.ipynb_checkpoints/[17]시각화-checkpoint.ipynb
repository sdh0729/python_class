```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline   # plt객체를 통해 그래프를 show() 렌더링, 대신에 사용
```

```python
d = [7, 1, 0, 6, 8, 5, 2, 15]
plt.plot(d)   # x,y 축 값의 인수를 내부 지정
```

```python
# plt.plot(d)   -> 점과 점을 선으로 연결해 준다. plt.plot(x,y). 여러 개가 들어올 때는 plt.plot(x[i], y[i])

x= [0,1,2,3,4,]
y = [0,3,6,9,12]


plt.plot(x,y)   #x, y 축 값의 인수를 내부 지정
```

```python
# numpy로 연동
# plt.plot(d)   -> 점과 점을 선으로 연결해 준다.

x= [0,1,2,3,4,]
array_X = np.array(x)
y = [0,3,6,9,12]
array_Y = np.array(y)
plt.plot(array_X, array_Y)
```

```python
# 여러 개의 선으로 표시
data = [0,1,4,9,16]
x = [0,1,2,3,4]
y = [0,1,2,3,4]
plt.plot(x,y)
plt.plot(data)
```

```python
# 여러 개의 선으로 표시   속성을 확인하자.  linestyle, color, marker
data = [0,1,4,9,16]
x = [0,1,2,3,4]
y = [0,1,2,3,4]

plt.plot(x,y, linestyle = '--', color = 'blue', marker='o', label = 'linear')
plt.plot(data, linestyle = ':', color = 'green', marker='*', label = 'quad')

plt.legend()


plt.title('My First G')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python
# 곡선
x = np.arange(-np.pi, np.pi, 0.1)

plt.plot(x, np.cos(x), marker = 'o')   # 영상 -> 주파수 / 이미지 분류 -> 주파수 확인
plt.plot(x, np.sin(x))
```
