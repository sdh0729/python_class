```python
import bs4
from bs4 import BeautifulSoup
import requests
from urllib.error import URLError, HTTPError
import urllib.request
import pandas as pd
```

```python
url = 'https://www.melon.com/chart/index.htm'
headers = {'User-Agent':'Chrome/66.0.3359.181'}
req = urllib.request.Request(url, headers=headers)
html = urllib.request.urlopen(req)
bs_obj = BeautifulSoup(html,'html.parser')
```

```python
t_list = bs_obj.find_all('div',{'class':'ellipsis rank01'})
s_list = bs_obj.findAll('div',{'class':'ellipsis rank02'})
```

```python
t_list
```

```python
title=[]
song=[]
```

```python
for li in t_list:
    a_tag = li.find('a')
    title.append(a_tag.text)
for li in s_list:
    a_tag = li.find('a')
    song.append(a_tag.text)
```

```python
len(title)
```

```python
len(song)
```

```python
df = pd.DataFrame({'곡 제목': title, '가수': song})
df
```

```python
df.insert(0,'순위',rank)
```
