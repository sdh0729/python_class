# 포털 사이트 크롤링

* 소스내에서 특정 문자열(data)을 지칭하는 선택자 얻기
    * 크롬 개발자도구 사용
* 전체 코드에서 수집하려고 하는 데이터(태그)의 위치를 찾고
    * 태그를 파싱한 후 필요데이터 추출


```python
from urllib.request import urlopen # 서버 요청/응답 패키지
import bs4 #파싱 패키지
import pandas as pd
```

```python
# 네이버 사이트의 기본 메뉴 문구 추출
url = 'https://www.naver.com'

# url로 요청 후 응답
html =urlopen(url)

#파서 객체 생성 = bs4 객체로 변환
bs_obj = bs4.BeautifulSoup(html,'html.parser')
```

```python
bs_obj
```

```python
test_menu = bs_obj.find('div',{'id':'gnb'})
test_menu
```

```python
#NM_FAVORITE > div.group_nav > ul.list_nav.type_fix
ul = bs_obj.find('ul',{'class':'list_nav type_fix'})
ul # 태그 객체
```

```python
ul_test = bs_obj.select('#NM_FAVORITE > div.group_nav > ul.list_nav.type_fix')
ul_test # list 형태로 반환
```

```python
type(ul) #bs4.element.Tag
# 문자열이아님
# 반복순환에 사용 가능
for u in ul:
    print(u)
```

```python
lis = ul.findAll('li')# 리스트 반환
for li in lis:
    print(li.find('a')['href'])
    print(li.text)
```

### 네이버 전체 메뉴 크롤링

```python
#네이버 기본 메뉴는 ul > li.nav_item 형식으로 되어 있음
lis = bs_obj.findAll('li',{'class':'nav_item'})
len(lis) # class 속성값이 nav_item인 li 태그 추출 : 16개
```

```python
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text,':',a_tag['href'])
```

### 수집한 데이터를 csv로 저장

1. 항목별로 list에 저장
2. 항목들을 dict로 구성
3. dict를 데이터프레임으로 생성
4. 데이터 프레임을 csv로 저장

```python
# 네이버 메뉴 정보 저장
# 빈 리스트 생성
menu = []
url = []
```

```python
for li in lis:
    a_tag = li.find('a')
    menu.append(a_tag.text)
    url.append(a_tag['href'])
```

```python
menu
```

```python
url
```

```python
df = pd.DataFrame({'메뉴': menu, 'URL': url})
df
```

```python
#df.to_csv 
df.to_csv('naver_menu.csv',encoding = 'utf-8')
```

## 네이버 뉴스 크롤링

* 단, 로그인 후 구독을 추가하면 구독한 언론사들의 뉴스가 나옴
* 헤드라인 뉴스는 표면적으로는 제공되지 않는다

```python
# 네이버 뉴스 크롤링
# url은 기본 url 부터 사용
url = 'https://news.naver.com'
html = urlopen(url)
```

```python
# bs4 객체 생성
bs_obj = bs4.BeautifulSoup(html,'html.parser')
```

## 네이버 뉴스섹션 메뉴와 섹션별 url 추출

```python
url = 'https://news.naver.com'

html = urlopen(url)

bs_obj = bs4.BeautifulSoup(html,'html.parser')
```

```python
# 네이버 뉴스 섹션 메뉴 태그 확인(개발자도구)
# selector가 너무 길어서 유용하지 않아 보임
# 직접 확인한 태그와 클래스 속성 사용
# ul 태그의 class : NLnb_menu_list

uls = bs_obj.findAll('ul',{'class':'Nlnb_menu_list'})
print(len(uls))
ul = bs_obj.find('ul',{'class':'Nlnb_menu_list'})
ul
```

```python
lis = ul.findAll('li')
len(lis)
```

```python
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text,':',a_tag['href'])
```

```python
# li 태그를 이용해서 추출
lis = bs_obj.findAll('li',{'class':'Nlist_item'})
len(lis)
```

```python
for li in lis:
    a_tag = li.find('a')
    print(a_tag.text)
    print(a_tag['href'])
```

```python
#수집 데이터 df로 구성 후 저장
section =[]
link= []
```

```python
for li in lis:
    a_tag = li.find('a')
    section.append(a_tag.text)
    link.append(a_tag['href'])
```

```python
col_dict = {'section':section, 'link':link}
news_section_df = pd.DataFrame(col_dict)
news_section_df
```

```python
# 저장
news_section_df.to_csv('naver_news_section.csv',encoding='utf-8')
```
