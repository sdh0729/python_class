```python
from selenium import webdriver
from bs4 import BeautifulSoup
import bs4
import pandas as pd
import time
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import platform
```

```python
driver = webdriver.Chrome("chromedriver")
url = 'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube'
driver.get(url)
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
```

```python
rank_info = []
rank = 1
lis = soup.select('form > table > tbody > tr')
for li in lis:
    category = li.select('p.category')[0].text.strip()
    title = li.select('h1 > a')[0].text.strip()
    subscribe = li.select('.subscriber_cnt')[0].text
    view = li.select('.view_cnt')[0].text
    video = li.select('.video_cnt')[0].text
    rank_info.append(['Youtube',rank,category,title,subscribe,view,video])
    rank = rank +1
```

```python
columns = ['회사', '순위',  '카테고리', '타이틀', '구독자수', '조회수','비디오수']
df = pd.DataFrame(rank_info, columns=columns)
```

```python
df = pd.read_csv('youtube_rank.csv')
df.head
```

```python
df['replaced_subscriber'] = df['subscriber'].str.replace('만', '0000')
df['replaced_subscriber'] = df['replaced_subscriber'].astype('int')
```

```python
df
```

```python
pivot_df = df.pivot_table(index = 'Category', values = 'replaced_subscriber',aggfunc=['sum', 'count'])
pivot_df
```

```python
pivot_df.columns = ['subscriber_sum', 'category_count']
```

```python
pivot_df = pivot_df.reset_index()
pivot_df = pivot_df.sort_values(by='subscriber_sum', ascending = False)
pivot_df
```

```python
plt.rc('font',family='AppleGothic')
plt.figure(figsize=(30,10))
plt.pie(pivot_df['subscriber_sum'], labels=pivot_df['Category'],autopct='%.1f%%')
plt.show
```

```python
pivot_df = pivot_df.sort_values(by='category_count', ascending = False)
plt.rc('font',family='AppleGothic')
plt.figure(figsize=(30,10))
plt.pie(pivot_df['category_count'], labels=pivot_df['Category'],autopct='%.1f%%')
plt.show
```
