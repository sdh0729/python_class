```python
for y in range(1880,2016):
    filename = 'C:\ds_work\data\EXAM01\yob{0}.txt'.format(y)
    yobFile = open(filename,'r')
    babies = yobFile.readlines()
    
    boys = [baby for baby in babies if baby.split(',')[1]=='M']
    girls = [baby for baby in babies if baby.split(',')[1]=='F']

    print("{0}년도 신생아 : 남아 {1}명, 여아{2}명".format(y,len(boys),len(girls)))
    yobFile.close()
```

```python
for y in range(1880,2016):
    df = pd.read_csv(os.path.join(path, 'yob{}.txt'.format(y)), 
                     sep=',', 
                     names=['name','sex','babies'], 
                     header=None)
    df.to_csv(path_or_buf='C:\ds_work\data'+'yob{}.csv', index=False)
```
