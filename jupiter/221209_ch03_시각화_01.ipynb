## 여러 개의 선 그어보기

```python
import numpy as np
t = np.arange(0., 5., 0.2) # 0부터 5사이의 (끝X) 0.2 간격으로 나열된 배열
t
```

```python
# 1. 여러 개의 선 그리기 (하나의 plot 사용)
plt.title('라인 플롯에서 여러 개의 선') 
plt.plot(
    t, t, 'r--', # x1, y1, 스타일
    t, t ** 2 - 1, 'bs:', # x2, y2, 스타일
    t, t ** 3 - 3, 'g^-' # x3, y3, 스타일
    # ... 
)
plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
plt.title('여러 개의 plot 명령을 한 그림에 겹쳐서 표현')
plt.plot((np.arange(4) + 1) ** 2) # 1
plt.plot([9, 16, 4, 1]) # 2
plt.show() # 이걸 기준으로 그려짐
```

```python
# 여러 개의 플롯이 겹쳤을 경우 라벨 표시
X = np.linspace(-np.pi, np.pi, 256)
X
```

```python
# 코사인, 사인 곡선
C, S = np.cos(X), np.sin(X)
C, S
```

```python
plt.title('삼각함수 plot')
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
plt.title('legend를 표시한 plot')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
# loc = 0~10
plt.legend(loc=2) # 상단왼쪽 
plt.show()
```

### `subplot`
* plot 명령을 여러 번 실행 => 같은 그래프 내부에 여러 직선 표시
* subplot : 행과 열 단위로 묶어서 여러 데이터를 병행해서 표시

```python
import numpy as np
```

```python
# 위아래 행 배치
# 행, 열, 서브플롯의 위치
plt.subplot(2, 1, 1) # 2행 중에 1행에서 그려질 플롯
plt.plot(np.arange(5), np.arange(5) ** 2)
plt.title('차트1')
plt.subplot(2, 1, 2) # 2행 중에 2행에서 그려질 플롯
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title('차트2')
# 겹치는 문제
plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show()
```

```python
# 양옆 열 배치
# 행, 열, 서브플롯의 위치
plt.subplot(1, 2, 1) # 2행 중에 1행에서 그려질 플롯
plt.plot(np.arange(5), np.arange(5) ** 2)
plt.title('차트1')
plt.subplot(1, 2, 2) # 2행 중에 2행에서 그려질 플롯
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title('차트2')
# 겹치는 문제
plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show()
```

```python
# 2x2 플롯 배치
# 행, 열, 서브플롯의 위치
# plt.subplot(2, 2, 1)
plt.subplot(221)
plt.plot(np.arange(5), np.arange(5) ** 2)
plt.grid(True)
plt.title('차트1')
# plt.subplot(2, 2, 2)
plt.subplot(222)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.xlabel('x축')
plt.title('차트2')
# plt.subplot(2, 2, 3)
plt.subplot(223)
plt.plot(np.arange(10), np.arange(10) ** 3, 'g-')
plt.ylabel('y축')
plt.title('차트3')
# plt.subplot(2, 2, 4)
plt.subplot(224)
plt.plot(np.arange(10), np.arange(10) ** 3, 'b^-')
plt.title('차트4')
# 겹치는 문제
plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show()
```

## 다양한 차트 활용
* `lineplot` 선 차트
* 그외에 여러 가지 차트가 존재

```python
import matplotlib.pyplot as plt

y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다'] # x, xticks-label 갯수가 같아야함
plt.title('바(bar) 차트')
# xticks 설정 -> x축을 깔끔하게
plt.xticks(x, xlabel) # x축에 개별 값에다가 이름 붙여줌
plt.yticks(sorted(y)) # y축에다가 기준에 되는 이름 붙여주기
plt.bar(x, y)
plt.show()
```

```python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다'] # x, xticks-label 갯수가 같아야함
plt.title('바(bar) 차트')
# xticks 설정 -> x축을 깔끔하게
plt.yticks(x, xlabel) # x축에 개별 값에다가 이름 붙여줌
plt.xticks(sorted(y)) # y축에다가 기준에 되는 이름 붙여주기
plt.barh(x, y)
plt.show()
```

```python
np.random.seed(1209)
people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(4)
performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)

plt.title('Bar-h Chart')
plt.grid(True)
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x축')
plt.show()
```

### 파이 차트
* 원형의 파이 모양 그래프

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```
